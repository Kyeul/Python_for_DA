## Matplotlib으로 데이터 시각화하기

Matplotlib이란 파이썬에서 데이터를 시각화하는 라이브러리이다. 과학 및 공학 연산을 위한 소프트웨어인 __MATLAB__의 시각화 기능을 모델링하여 만들어졌다. Matplotlib을 사용하면 짧은 코드만으로 다양한 종류의 그래프를 효과적으로 만들 수 있다.

더욱 자세한 정보는 __matplotlib__의 홈페이지(http://matplotlib.org/)를 통해 얻을 수 있다.

matplotlib의 그래프 기능을 이용하려면 다음의 서브모듈을 불러와야 한다.

````python
import matplotlib.pyplot as plt
````

위와 같이 matplotlib의 서브모듈을 불러오면 plt를 이용해 사용할 수 있다.   
이제 matplotlib에서 제공하는 그래프 기능을 `plt.그래프_함수()`형식으로 사용할 수 있다.

matplotlib의 그래프를 실행하면 기본적으로 코드 결과 출력 부분에 나온다.   
만약 별도의 팝업창에 그래프를 출력하려면 다음을 먼저 실행해야 한다.

````python
%matplotlib qt
````

그래프를 다시 코드 결과 출력 부분에 출력하려면 다음을 실행한다.

````python
%matplotlib inline
````

### 선 그래프
#### 기본적인 선 그래프 그리기

Matplotlib에서는 다음과 같은 형식으로 2차원 선 그래프를 그린다.


``` plt.plot([x,] y [,fmt]) ```

x와 y는 각각 x축과 y축 좌표 값을 의미한다.   
즉, x와 y는 각각 2차원 좌표 집합 $\{(x_0, y_0), (x_1, y_1), (x_2, y_2), \ldots,(x_n, y_n)\}$에서 x축 좌표의 요소만 모아서 만든 sequence data $[x_0, x_1, x_2, \ldots, x_n]$과 y축 좌표의 element만 모아서 만든 sequence data $ [y_0, y_1, y_2, \ldots, y_n] $을 의미한다. 다시 말해 $ x=[x_0, x_1, x_2, \ldots, x_n] $이고, $y=[y_0, y_1, y_2, \ldots, y_n]$이다. 여기서 x와 y의 sequence의 길이는 같아야 한다. __x와 y data 중에서 x는 생략할 수 있다.__ x가 없다면 x는 0부터 y의 개수만큼 1씩 증가하는 값으로 자동 할당된다.  

__fmt__는 __format string__으로 다양한 형식으로 그래프를 그릴 수 있는 옵션이다. 

<연습1>

````python
data1 = [10, 14, 19, 20, 21] #숫자로 이루어진 sequence data 생성
plt.plot(data1) # y값에 data1을 이용해 선 그래프를 생성
````

y축 데이터가 data1인 그래프가 그려졌다. x축은 별도의 데이터를 넣지 않아 0부터 1씩 증가하는 값이 자동으로 할당됐다.   
결과를 살펴보면 5개의 $(x, y)$ 좌푯값인 $\{(0, 10), (1, 14), (2, 19), (3, 20), (4, 21)\}$가 그려졌음을 확인할 수 있다.   
__`plt()`__로 데이터를 시각화할 때 특별한 옵션을 지정하지 않으면 기본적으로 선 그래프가 그려진다.

<! 그래프의 객체 정보 제거하기>  

객체 정보 없이 그래프만 출력하려면 `그래프_함수()`다음 line에 `show()`를 함께 적으면 된다.

````python
plt.plot(data1)
plt.show()
````

#### x값과 y값이 모두 있는 2차원 선 그래프 그리기

<연습2>

````python
import numpy as np

x = np.arange(-4.5, 5, 0.5) # 배열 x생성. [-4.5, 5) 범위에서 0.5씩 증가하는 데이터
y = 2*x**2 # 배열 x에 대응하는 배열 y생성 => 2x^2
[x, y]
````

x값과 y값으로 그래프 그리기

````python
plt.plot(x, y)
plt.show()
````

plot()을 이용해 $y=2x^2$ 관계가 있는 x와 y 데이터를 시각화했다. 

### 한 figure에 여러가지 그래프 그리기

<연습1>과 <연습2>에서는 하나의 선 그래프를 그렸다. 이번에는 여려 개의 데이터를 효과적으로 시작화하는 방법을 살펴보자

하나의 그래프 figure에 여러 개의 데이터를 선 그래프로 표시하는 방법은 다음과 같다.

```
plt.plot([x1,] y1 [,fmt1])
plt.plot([x2,] y2 [,fmt2])
plt.plot([x3,] y3 [,fmt3])
...
plt.plot([xn,] yn [,fmtn])
```
위와 같이 __연속해서 plot()__을 수행하면 하나의 figure에 여러 그래프를 그릴 수 있다.  
위와 같은 방법으로 다음과 같은 표기법도 사용할 수 있다.

``` 
plt.plot(x1, y1[, fmt1], x2, y2 [, fmt2], ... ,xn,yn[, fmtn])
```
위 형식의 경우 x와 y 데이터가 쌍을 이루어야 하므로 x 데이터를 생략할 수 없다.

<연습3>

````python
x = x #<연습2>에서 생성한 x를 재사용
y1 = y #<연습2>에서 생성한 y를 재사용
y2 = 5*x + 30 # 5x+30
y3 = 4*x**2 + 10 #4x^2+10
````

````python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
````

위의 결과를 보면 x에 대한 데이터 y1, y2, y3가 각각 다른 색의 선 그래프로 그려진 것을 볼 수 있다.  
추후에 각 데이터의 선 색을 지정하는 옵션도 살표볼 예정이다.

한 줄로 작성하는 코드는 다음과 같다

````python
plt.plot(x,y1, x,y2, x,y3)
plt.show()
````

#### 여러의 figure에 그래프 그리기
하나의 그래프 창에 여러 개의 그래프를 그리는 대신 새로운 그래프 창에 그리려면 `'그래프_함수()'`를 실행하기 전에 `plt.figure()`로 새로운 그래프 창을 생성해야 한다.

<연습4>

````python
plt.plot(x, y1)

plt.figure()
plt.plot(x, y2)

plt.show()
````

위에서 볼 수 있듯이 `plt.figure()`를 기준으로 그래프 figure가 새로 생성됨을 확인했다.

다음의 코드로 그래프 창의 번호를 명시적으로 줄 수도 있다.  

`plt.figure(n)`
n에 번호를 넣으면 지정된 번호의 그래프 figure가 생성된다.

현재 그래프 figure에 그래프를 모두 지우기 위해서는 `plt.clf()`를 사용하면 된다.

````python
plt.plot(x, y1)
plt.clf()
plt.show()
````

<연습5>
두 개의 데이터 쌍 $(x, y_1), (x, y_2)$를 지정된 번호의 그래프 figure에 그래프로 그려보자.  
여기서 x에 대한 y1, y2 데이터 생성을 위한 수식은 각각 $y_1=x^2-2, y_2=20\cos(x)^2$이다.

````python
x = np.arange(-5, 5, 0.1) # [-5, 5)까지 범위에서 0.1간격으로 데이터 생성
y1 = x**2-2
y2 = 20*np.cos(x)**2

plt.figure(1) # 1번 figure 생성
plt.plot(x, y1)

plt.figure(2) # 2번 figure 생성
plt.plot(x, y2)

plt.figure(1) # 1번 figure를 지정함
plt.plot(x, y2) #지정된 그래프 창에 그래프를 그림

plt.figure(2) # 2번 figure를 지정함
plt.clf() #지정된 figure의 그래프를 삭제
plt.plot(x, y1) #지정된 figure에 그래프를 그림

plt.show()
````
